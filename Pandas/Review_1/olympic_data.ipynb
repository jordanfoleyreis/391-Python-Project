## Adding & Removing Columns

```python
db['height_category'] = db['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
```

```python
db
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return "Lightweight"
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return "Middleweight"
    else:
        return "Heavyweight"
    
db['Category'] = db.apply(categorize_athlete, axis = 1)
```
