## Aggregating Data

```python
coffee.groupby(['Coffee Type'])['Units Sold'].agg({'Units Sold': 'sum', 'price': 'mean'}) # can also do mean(), sum()
```

```python
coffee.groupby(['Coffee Type', 'Day'])['Units Sold'].agg({'Units Sold': 'sum', 'price': 'mean'}) # can also do mean(), sum()
```

```python
pivot = coffee.pivot(columns = 'Coffee Type', index = 'Day', values = 'revenue')
pivot.sum()
#pivot.loc...
```

```python
coffee['Yesterday_revenue']= coffee['revenue'].shift()
```

```python
coffee['cum_rev']= coffee['revenue'].cumsum()
```

```python
lattee= coffee[coffee['Coffee Type']== 'Latte'].copy()
lattee['3 Day']= lattee['Units Sold'].rolling(3).sum()
```
