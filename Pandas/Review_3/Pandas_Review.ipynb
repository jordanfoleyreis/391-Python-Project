# Data Cleaning 

```python
df.isnull().sum() #goes by each column and gives total of null values 
#You can either get rid of all rows or insert values
```

```python
#df.dropna 
#Drops all na values from each row, but you didn't use inplace = true so it does not stay in the orginal data fram e
```

```python
df.dropna(how = "any") #Drops all na values no inplace = true, Any row that had any na value
#can all take how = "All" #only if all values in a row are na if not then they are no deleted
```

```python
df["Age"].fillna(df["Age"].mean())
#df["Age"].fillna(df["Age"].mean(),inplace = True)
```

```python
#df.fillna(0) #NOOB METHOD 
```

```python
df["Salary"].fillna(df["Salary"].median())
```

```python
df["Age"].fillna(method = "ffill")
#ffill = forward fill, you move from top to bottom 
```

```python
df["Age"].fillna(method = "bfill")
#bfill = backward fill, you fill in the value from buttom to top 
```

```python
df["Name"]=df["Name"].replace("Charlie", "Rose") #replace values that you wish to change
df
```

# Joins and Merges



```python
department_info = {
    "Dept": ["HR", "IT", "Finance"],
    "Location": ["New York", "San Francisco", "Chicago"],
    "Manager": ["Laura", "Steve", "Nina"]
}

df2 = pd.DataFrame(department_info)
df2
```

```python
# Concat

df_concat= pd.concat([df, df2])
df_concat
#Allows for full outher join
```

```python
pd.concat([df, df2], axis = 1)
#Join on df2 # review axis
```

```python
pd.merge(df, df2, on = "Dept")
```
